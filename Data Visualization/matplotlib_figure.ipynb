```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,9,20)
y = x**3

figure = plt.figure()                      # empty figure
axes = figure.add_axes([0.2,0.2,0.9,0.9])  # axes in the figure
axes.plot(x,y,"b")
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes.set_title("Cube")

plt.show()
```

```python
x = np.linspace(-10,9,20)
y = x**3
z = x**2

figure = plt.figure()                      # empty figure

axes = figure.add_axes([0.1,0.1,0.9,0.9])  # axes in the figure
axes.plot(x,y,"b")
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes.set_title("Cube")

axes2 = figure.add_axes([0.15,0.7,0.25,0.25])
axes2.plot(x,z,"r")
axes2.set_xlabel("X axis")
axes2.set_ylabel("Y axis")
axes2.set_title("Square")

plt.show()
```

```python
x = np.linspace(-10,9,20)
y = x**3
z = x**2

figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(x,y,label="cube")
axes.plot(x,z,label="square")
axes.legend(loc=4)
plt.show()
```

```python
x = np.linspace(-10,9,20)
y = x**3
z = x**2

#fgr , axs = plt.subplots(2,1)
fgr , axs = plt.subplots(nrows = 2, ncols=1)
axs[0].plot(x,y)
axs[0].set_title("cube")
axs[1].plot(x,z)
axs[1].set_title("square")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-10,9,20)
y = x**3
z = x**2

#fgr , axs = plt.subplots(2,1)
fgr , axs = plt.subplots(nrows = 2, ncols=1 , figsize =(4,4))
axs[0].plot(x,y)
axs[0].set_title("cube")
axs[1].plot(x,z)
axs[1].set_title("square")
plt.tight_layout()
fgr.savefig("png1.png")
fgr.savefig("png1.pdf")

plt.show()
```
