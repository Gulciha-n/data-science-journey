# dropna

NaN (Not a Number) ,
None (missing values)

```python
df.dropna()    #default axis=0 (row)
```

```python
df.dropna(axis=1)    # column
```

```python
df
```

```python
df.dropna(how="any")   #bir satır içerisinde nan değeri olan bir tane bile değer varsa siler
```

```python
df.dropna(how="all")  #tüm değerleri nan olan satırı siler
```

# subset
belli column'ları seçmek için "subset" kullanılır

```python
df.dropna(subset=["column1","column2"] , how = "all")
#column1 ve column2'de ikisinde de nan olan değerler silinir
#bir tanesinde nan diğeri değilse o satır silinmez
```

```python
df.dropna(subset=["column1","column2"] , how = "any")
#column1 ve column2'de iki satırda birinde bile nan varsa o satır silinir
```

# thresh  = n
nan dısında n tane değer varsa o satırı veya sutunu silme anlamına gelir,

axis=0 veya belirtilmemiş ise satırı etkiler,

axis=1 ise sütunu etkiler


```python
df.dropna(thresh = 2)
#en az iki veri varsa o satırı silme, o satırda ikiden az değer varsa o satır silinir
```

```python
df.dropna(thresh = 4)  #en az 4 değer varsa o satırı silme
```

```python
df.dropna(thresh = 3)  #en az sayıda normal veri (her satırda bir tane nan kabul edilebilir)
```

# Filling the expected values with the average value of the table

```python
print(df.sum())          #her columndaki sayıların toplamı
print(df.sum().sum())    #df'deki tüm kolonların toplamı
print(df.size)           #eleman sayısı
print(df.isnull().sum()) #null olan eleman sayısı
print(df.isnull().sum().sum()) #null olan toplam eleman sayısı


def average(df):
  summation = df.sum().sum()
  item_number = (df.size) - (df.isnull().sum().sum())
  avr = summation / item_number
  return avr

result = average(df)
print(f"Average : {result}")

df.fillna(value = average(df))   #nan olan değerleri tablo ortalamasına göre doldurduk
```
