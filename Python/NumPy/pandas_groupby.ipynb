```python
import pandas as pd
import numpy as np
```

```python
employees = {
    'Employee': ['John Smith', 'David Johnson', 'Michael Brown',
                 'James Davis', 'Robert Miller', 'William Wilson', 'Charles Moore'],

    'Department': ['Human Resources', 'Information Technology', 'Accounting', 'Human Resources',
                   'Information Technology', 'Accounting', 'Human Resources'],

    'Age': [30, 25, 45, 50, 23, 34, 42],

    'District': ['Kadıköy','Tuzla','Maltepe','Tuzla','Maltepe','Tuzla','Kadıköy'],

    'Salary': [5000, 3000, 4000, 3500, 2750, 6500, 4500]
}

df = pd.DataFrame(employees)

print(df)
```

```python
df.groupby('District').groups
```

```python
df.groupby(["District", "Department"]).groups
```

```python
for name,group in  df.groupby("District"):   #birinci elemanının grup anahtarı
  print(name)                                #ikinci elemanının o gruba ait satırlar
  print(group)
  print()
```

#aggregate function

**aggregate function ile birden fazla parametreyi ayni anda kullanabiliyoruz**

```python
df.groupby("District")["Salary"].max()
```

```python
df.groupby("District")["Salary"].agg(['max', 'min', 'mean'])        #using pandas for agg function
```

```python
df.groupby("District")["Salary"].agg([np.max , np.min, np.mean])    #using numpy for agg function
```

```python
df.groupby("District")["Salary"].agg([np.max , np.min, np.mean]).loc["Kadıköy"]
```

```python
df.groupby("District")["Salary"].agg([np.max , np.min, np.mean]).loc["Tuzla"]
```

```python
df.groupby("District")["Salary"].agg([np.max , np.min, np.mean]).loc["Maltepe"]
```

```python
df.groupby("District")["Salary"].agg([np.max , np.min, np.mean]).loc[ : , "mean"]  # all lines - mean columns
```

```python
df.groupby("District")["Salary"].agg([np.max , np.min, np.mean]).loc[ :,"max"]
```
